### As colunas da pesquisa possuem nomes pouco intuitivos. Renomeando as variáveis mais relevantes para facilitar a interpretação dos resultados.

```python
df = df.rename(columns={
    '1.a_idade': 'idade',
    '1.b_genero': 'genero',
    '1.c_cor/raca/etnia': 'raca',
    '1.i_estado_onde_mora': 'estado',
    '1.i.2_regiao_onde_mora': 'regiao',
    '1.l_nivel_de_ensino': 'escolaridade',
    '2.a_situação_de_trabalho': 'situacao_trabalho',
    '2.b_setor': 'setor',
    '2.f_cargo_atual': 'cargo',
    '2.g_nivel': 'senioridade',
    '2.h_faixa_salarial': 'faixa_salarial',
    '2.i_tempo_de_experiencia_em_dados': 'exp_dados',
    '2.j_tempo_de_experiencia_em_ti': 'exp_ti',
    '2.k_satisfeito_atualmente': 'satisfacao',
    '2.r_modelo_de_trabalho_atual': 'modelo_trabalho'
})
```

```python
#Colunas Relevantes para o Dataset
colunas_relevantes = [
    'idade',
    'genero',
    'raca',
    'estado',
    'regiao',
    'escolaridade',
    'situacao_trabalho',
    'setor',
    'cargo',
    'senioridade',
    'faixa_salarial',
    'exp_dados',
    'exp_ti',
    'satisfacao',
    'modelo_trabalho'
]
```

```python
df = df[colunas_relevantes]
df
```

## Conversão da variável salário

A variável salário está armazenada em formato textual.
Para permitir cálculos estatísticos, cada faixa salarial foi convertida para seu ponto médio.

Exemplo:

R$ 4.001 a R$ 6.000 → R$ 5.000

Essa decisão introduz pequena imprecisão, porém permite análises quantitativas.

```python
mapa_salario = {
    'Menos de R$ 1.000/mês':500,
    'de R$ 1.001/mês a R$ 2.000/mês':1500,
    'de R$ 2.001/mês a R$ 3.000/mês':2500,
    'de R$ 3.001/mês a R$ 4.000/mês':3500,
    'de R$ 4.001/mês a R$ 6.000/mês':5000,
    'de R$ 6.001/mês a R$ 8.000/mês':7000,
    'de R$ 8.001/mês a R$ 12.000/mês':10000,
    'de R$ 12.001/mês a R$ 16.000/mês':14000,
    'de R$ 16.001/mês a R$ 20.000/mês':18000,
    'de R$ 20.001/mês a R$ 25.000/mês':22500,
    'de R$ 25.001/mês a R$ 30.000/mês':27500,
    'Acima de R$ 30.000/mês':35000
}

df['salario_num'] = df['faixa_salarial'].map(mapa_salario)
df
```

## Conversão da variavel exp_dados

```python
# Conversão da experiência em dados

mapa_exp_dados = {
    'Não tenho experiência na área de dados': 0,
    'Menos de 1 ano': 0.5,
    'de 1 a 2 anos': 1.5,
    'de 3 a 4 anos': 3.5,
    'de 5 a 6 anos': 5.5,
    'de 7 a 10 anos': 8.5,
    'Mais de 10 anos': 12
}

df['exp_dados_num'] = df['exp_dados'].map(mapa_exp_dados)
```

```python
# Conversão da experiência em TI

mapa_exp_ti = {
    'Não tive experiência na área de TI/Engenharia de Software antes de começar a trabalhar na área de dados': 0,
    'Menos de 1 ano': 0.5,
    'de 1 a 2 anos': 1.5,
    'de 3 a 4 anos': 3.5,
    'de 5 a 6 anos': 5.5,
    'de 7 a 10 anos': 8.5,
    'Mais de 10 anos': 12
}

df['exp_ti_num'] = df['exp_ti'].map(mapa_exp_ti)
```

```python
df[['exp_dados', 'exp_dados_num',
    'exp_ti', 'exp_ti_num']].head()
```

Campos de gênero e raça não foram imputados.

Segundo a documentação da pesquisa, a ausência de resposta nesses campos pode representar uma escolha do participante.

```python
df['cargo'] = df['cargo'].fillna('Não informado')
```

```python
df['cargo'] = df['cargo'].str.lower()

mapa_cargos = {
    'data analyst':'analista de dados',
    'analista de dados jr':'analista de dados',
    'analista de dados pleno':'analista de dados',
    'analista de dados senior':'analista de dados'
}

df['cargo'] = df['cargo'].replace(mapa_cargos)
```

Correlação

```python
corr = df[['salario_num',
           'exp_dados_num',
           'exp_ti_num']].corr(method='spearman')

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='Blues',
    fmt='.2f'
)
    
plt.title('Correlação entre salário e experiência')
plt.show()
```

5. Síntese Visual

## Qual fator mais influencia o salário?

A senioridade apresentou a maior diferença salarial entre os grupos analisados.

Profissionais seniores apresentaram remuneração substancialmente superior aos níveis júnior e pleno, indicando que experiência profissional é um dos principais determinantes salariais no mercado brasileiro de dados.

```python
media_senioridade = (
    df.groupby('senioridade')['salario_num']
    .mean()
    .sort_values()
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=media_senioridade.index,
    y=media_senioridade.values
)

plt.title('Fator com Maior Impacto no Salário: Senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário Médio')

plt.show()
```
